### Marketing Campaing Analysis

```python
campaing_data = data[['AcceptedCmp1',
                 'AcceptedCmp2',
                 'AcceptedCmp3',
                 'AcceptedCmp4',
                 'AcceptedCmp5',
                 'Response',
                 'Complain',
                 'AcceptedCmpOverall']]

campaing_data
```

```python
more_than_one = campaing_data[campaing_data['AcceptedCmpOverall']>1]
complain_without_accept = campaing_data[(campaing_data['AcceptedCmpOverall'] == 0 ) & (campaing_data['Complain'] > 0)]
accepted_once_and_complained = campaing_data[(campaing_data['AcceptedCmpOverall'] == 1) & (campaing_data['Complain'] > 0)]
accepted_more_and_complained = campaing_data[(campaing_data['AcceptedCmpOverall'] > 1) & (campaing_data['Complain'] > 0)]
complained = campaing_data[(campaing_data['Complain'] > 0)]
accepted = campaing_data[(campaing_data['AcceptedCmpOverall']>0 )]
first_adopters = campaing_data[campaing_data['AcceptedCmp1']==1]
last_adopters = campaing_data[(campaing_data['Response']==1)]
conversion = round(((accepted.shape[0]+last_adopters.shape[0])/campaing_data.shape[0])*100)


print(f'For this company, there was a conversion of {conversion}%.\n'
      f'{first_adopters.shape[0]} converted in the first campaign and {last_adopters.shape[0]} converted in the last.\n'
      f'Having a total of {complained.shape[0]} complaints, {complain_without_accept.shape[0]} of which there was no acceptance of the campaign, '
      f'{accepted_once_and_complained.shape[0]} accepted once and {accepted_more_and_complained.shape[0]} accepted more than once.'
)
```

##### Comment: There is a discrepancy between "response" which should be whether the customer accepted the last campaign and "AcceptedCmpOverall" which is how many times the customer accepted the campaign at the end. In some data, it is indicated that the customer accepted the offer in the last campaign, but the "AcceptedCmpOverall" is set to zero. Therefore, I am considering that there are more than 5 campaigns, but "AcceptedCmpOverall" only considers the first 5 campaigns.

### Rebuilding original dataset

```python
df = data.copy()
df['marital_level'] = data[['marital_Divorced',
                               'marital_Married',
                               'marital_Single',
                               'marital_Together',
                               'marital_Widow']].idxmax(axis=1)
df['education_level'] = data[['education_2n Cycle',
                                'education_Basic',
                                'education_Graduation',
                                'education_Master',
                                'education_PhD']].idxmax(axis=1)

df.drop(['marital_Divorced',
         'marital_Married',
         'marital_Single',
         'marital_Together',
         'marital_Widow',
         'education_2n Cycle',
         'education_Basic',
         'education_Graduation',
         'education_Master',
         'education_PhD'], axis=1, inplace=True)


df.head()
```

### Creating classes of campaing acceptance:
##### a) best_buyer: Accepted the campaign more than once
##### b) first buyer: Accepted in the first campaign
##### c) hard_buyer: Accepted in the last campaign
##### d) regular_buyer: Accepted the offer between the first and last campaing
##### e) not_buyer: Didn't accepted the campaing
##### comment: If we have a best buyer that bought also in the first campaing, this client will be classified as best buyer

```python
df['buyer_status'] = 'regular_buyer'

df.loc[df['AcceptedCmpOverall'] > 1, 'buyer_status'] = 'best_buyer'
df.loc[(df['AcceptedCmp1'] == 1) & (df['AcceptedCmpOverall'] == 1), 'buyer_status'] = 'first_buyer'
df.loc[df['Response'] == 1, 'buyer_status'] = 'hard_buyer'
df.loc[(df['AcceptedCmpOverall'] == 0) & (df['Response'] == 0), 'buyer_status'] = 'not_buyer'
df.head()
```

```python
df.drop(['AcceptedCmp1',
        'AcceptedCmp2',
        'AcceptedCmp3',
        'AcceptedCmp4',
        'AcceptedCmp5',
        'Response',
        'AcceptedCmpOverall'], axis=1, inplace=True)
df.head()
```

```python
plt.figure(figsize=(10, 6))
df['buyer_status'].value_counts().plot(kind='bar', color='skyblue')

plt.title('Buyer status frequency')
plt.xlabel('Buyer status')
plt.ylabel('Frequency')
plt.show()
```
