```python
# Binary columns

bin_data = data[['AcceptedCmp3',
                 'AcceptedCmp4',
                 'AcceptedCmp5',
                 'AcceptedCmp1',
                 'AcceptedCmp2',
                 'Response',
                 'Complain',
                 'education_2n Cycle',
                 'education_Basic',
                 'education_Graduation',
                 'education_Master',
                 'education_PhD',
                 'marital_Divorced',
                 'marital_Married',
                 'marital_Single',
                 'marital_Together',
                 'marital_Widow'
                 ]]
```

```python
summary_data = pd.DataFrame(index=bin_data.columns, columns=['0', '1'])

for column in bin_data.columns:
    summary_data.loc[column, '0'] = bin_data[column].value_counts().get(0, 0)
    summary_data.loc[column, '1'] = bin_data[column].value_counts().get(1, 0)

summary_data = summary_data.astype(int)

fig, ax = plt.subplots(figsize=(10, 6))
summary_data.plot(kind='bar', stacked=False, ax=ax)
plt.title('Frequency of each class')
plt.xlabel('Columns')
plt.ylabel('Count')
plt.show()
```
